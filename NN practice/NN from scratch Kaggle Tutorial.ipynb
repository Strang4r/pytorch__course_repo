```python
df = pd.read_csv('breast-cancer-wisconsin.csv')
df.describe()
```

```python
df.drop('1000025',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['2.1']=df["2.1"].replace(2,0).replace(4,1)
df.info
```

```python
df['1.3']=df['1.3'].replace('?',0)

df['1.3']=df['1.3'].astype(int)
print(df.dtypes)
```

```python
column_names=list(df.columns)
```

```python
# df.info()
df.head(10)
```

```python
# X : X_train : All the features for training (excluding class)
# Y : Y_train : Actual output (Class)
from sklearn.model_selection import train_test_split

x=df.iloc[:,:]
y=df.iloc[:,-1]
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# scaling data to 0-1 scaler

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xTrain=scaler.fit_transform(xTrain)
xTest=scaler.fit_transform(xTest)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
#inititalizing NN
classifier = Sequential()
# units== количество скрытых слоев
#Activation== функция активации.
# input_dims == количество независимых переменных
classifier.add(Dense(units = 5,activation = 'relu', input_dim=10))
classifier.add(Dense(units=3, activation='relu'))
classifier.add(Dense(units=1, activation='sigmoid'))
```

```python
# setup model for learning
classifier.compile(optimizer="adam",loss='mean_squared_error',metrics=['accuracy'])
```

```python
# learning process
classifier.fit(xTrain,yTrain,batch_size=10,epochs=100)
```

```python
# predicting output
yPred_classifier=classifier.predict(xTest)
yPred=[1 if y>=0.5 else 0 for y in yPred_classifier]
print(yPred)
```

```python
# confusion matrix metric
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(yTest,yPred)
print(cm)
```
